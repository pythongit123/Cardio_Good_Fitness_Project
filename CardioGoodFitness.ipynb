# Data Analysis & Visualization

```python
# Set the style for the plots
sns.set(style="whitegrid")

# Distribution of Age
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], kde=True, bins=15, color='blue')
plt.title('Age Distribution of Customers')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

Age:The majority of customers being in their 20s to 30s.

```python
# Age Distribution by Gender
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Age', hue='Gender', multiple='stack', kde=True, palette='muted')
plt.title('Age Distribution by Gender')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

Male age 20s to 30s more active 

```python
# Age Distribution by Marital Status
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Age', hue='MaritalStatus', multiple='stack', kde=True, palette='pastel')
plt.title('Age Distribution by Marital Status')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# Distribution of Income
plt.figure(figsize=(10, 6))
sns.histplot(df['Income'], kde=True, bins=15, color='orange')
plt.title('Income Distribution of Customers')
plt.xlabel('Income')
plt.ylabel('Frequency')
plt.show()
```

Average 40000 to 60000 are more frequent. 

```python
# Boxplot of Miles by Product
plt.figure(figsize=(10, 6))
sns.boxplot(x='Product', y='Miles', data=df, palette='Set2')
plt.title('Miles Run by Product Type')
plt.xlabel('Product Type')
plt.ylabel('Miles')
plt.show()
```

TM798 showing using higher Miles uses within product type

```python
# Average Income by Product Type
plt.figure(figsize=(10, 6))
sns.barplot(x='Product', y='Income', data=df)
plt.title('Average Income by Product Type')
plt.xlabel('Product Type')
plt.ylabel('Average Income')
plt.show()
```
